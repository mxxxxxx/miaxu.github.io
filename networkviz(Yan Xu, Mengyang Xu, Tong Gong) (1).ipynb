```python
import json
import pandas as pd
import numpy as np
import bqplot
import ipywidgets
```

```python
films = json.load(open("shared/sp18-is590dv/data/star_wars/films.json"))
```

```python
films
```

```python
char_input = json.load(open("shared/sp18-is590dv/data/star_wars/people.json"))
characters = {}
for character in char_input:
    characters[character['pk']] = character['fields']
```

```python
chars = ["Boba Fett", "Yoda", "Jabba Desilijic Tiure", "Darth Vader", "Obi-Wan Kenobi",
         "Beru Whitesun lars", "Mon Mothma"]
```

```python
df = pd.DataFrame(columns = ['movies', 'characters'])
char_names = []
for film in films:
    film_chars = film['fields']['characters']
    film_name = film["fields"]["title"]
    char_names = []
    for character in film_chars:
        if characters[character]['name'] in chars: 
            character_name = characters[character]["name"]
            char_names.append(character_name)
    df = df.append({'movies': film_name, 'characters': char_names}, ignore_index=True)
```

```python
df
```

```python
nodes = [{"label": "Boba Fett"}, {"label": "Yoda"},{"label": "Jabba Desilijic Tiure"},
         {"label": "Darth Vader"},{"label": "Obi-Wan Kenobi"},{"label": "Beru Whitesun lars"},
        {"label": "Mon Mothma"}]
```

```python
for node in nodes:
    node["media"] = []
    node["shape"] = "circle"
```

```python
for film in films:
    film_chars = film["fields"]["characters"]
    film_characters = []
    for character in film_chars:
        film_characters.append(characters[character]["name"])
    for node in nodes:
        if node["label"] in film_characters:
            node["media"].append(film["fields"]["title"])
```

```python
nodes
```

```python
links_data = []
nnode = len(nodes)
for i in range(nnode):
    for j in range(i+1, nnode):
        for media in nodes[i]["media"]:
            if media in nodes[j]["media"]:
                links_data.append({'source': i, 'target': j})
                links_data.append({'source': j, 'target': i})
                break 
```

```python
links_data
```

```python
tooltip = bqplot.Tooltip(fields=["label", "media"])
graph = bqplot.Graph(node_data = nodes, link_data = links_data, tooltip = tooltip, link_type="line")

fig = bqplot.Figure(marks = [graph])
display(fig)
```

```python
tooltip = bqplot.Tooltip(fields=["label", "media"])
graph = bqplot.Graph(node_data=nodes, link_data=links_data, tooltip = tooltip, link_type="line")
fig = bqplot.Figure(marks = [graph])
def show_choose(obj, element):
    show = []
    idx = element['data']["index"]
    for links in links_data:
        if links["source"]==idx or links["target"]==idx:
            show.append(links)
    link_data=show
    graph1=bqplot.Graph(node_data=nodes, link_data=link_data, tooltip = tooltip, link_type="line")
    ipywidgets.link((graph1,"link_data"),(graph,"link_data"))
    

graph.on_element_click(show_choose)
display(fig)
```
